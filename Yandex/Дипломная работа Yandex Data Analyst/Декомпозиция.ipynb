# Декомпозиция

**Цель**: провести анализ клиентов регионального банка для выявления сегментов пользователей, склонных уходить из банка «Метанпром».

**Задача**: 


   - <u>Провести исследовательский анализ данных</u>:
   
  
        - Описать и визуализировать основные характеристики данных;
        
        - Определить значимые признаки оттока клиентов; 
        
        - Выделить границы значений, которые связаны с повышенным оттоком клиентов.


   - <u>Составить и сравнить "портреты" типичных клиентов банка</u>:
   
   
        - Сравнить характеристики клиентов, склонных к оттоку, с характеристиками клиентов, не склонных к оттоку;
        
        - Определить ключевые различия между этими группами.


   - <u>Сформулировать и проверить статистические гипотезы</u>:
   
   
        - Проверить гипотезу о различии дохода между клиентами, которые ушли, и теми, которые остались. (Нулевая гипотеза (H0): Средний доход клиентов, которые ушли, равен среднему доходу клиентов, которые остались. Альтернативная гипотеза (H1): Средний доход клиентов, которые ушли, не равен среднему доходу клиентов, которые остались.)
            
        - Проверить гипотезу о различии среднего возраста между клиентами, которые ушли, и теми, которые остались. (Нулевая гипотеза (H0): Средний возраст клиентов, которые ушли, равен среднему возрасту клиентов, которые остались. Альтернативная гипотеза (H1): Средний возраст клиентов, которые ушли, не равен среднему возрасту клиентов, которые остались.)
        
        - Проверить гипотезу о различии среднего баланса на счете между клиентами, которые ушли, и теми, которые остались.( Нулевая гипотеза (H0): Средний баланс на счете клиентов, которые ушли, равен среднему балансу на счете клиентов, которые остались.Альтернативная гипотеза (H1): Средний баланс на счете клиентов, которые ушли, не равен среднему балансу на счете клиентов, которые остались.)

  
  - <u>Подготовить вывод о клиентах банка и категории пользователей скланных уходить из него</u>.

**Этапы проекта**:

1. Обзор данных датасета


2. Предобработка данных на основе обзора данных:


        - Изменение регистра наименования колонок;

        - Проверка (корректировка) типа данных;

        - Проверка на наличие пропусков в данных;

        -  Проверка на наличие дубликатов (явных и неявных);

        -  Анализ пороговых значений и выбросов в данных.
        
        
        
3. Исследовательский анализ данных:

          1. Анализ основных харктеристик клиентов банка (город, возраст, гендер, доход клиентов, продукты, наличие кредитной карты, баллы собственности);

          2. Анализ значимых признаков по категории оттока клиентов (следует выбрать: возраст пользователей, доходы клиентов, город, продукты);

          3. Провести анализ значений по диаграме размаха по значимым признакам сгруппированных по признаку оттока.

4. Проверка стаистических гипотез:

       - Формулировка гипотез;
       -  Проведение тестирования;
       -  Анализ результатов.


5. Выводы


6. Подготовка презентации.
